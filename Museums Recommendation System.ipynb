```python
import pandas as pd

data = pd.read_csv('Tags.csv')

data.head()
```

```python
data['Tags']=data['Categories']+data['Keywords']+data['Description']+data['State_Name']

data.head()
```

### Removing Unnecesary colums and making new df

```python
museums=data[['Location_ID','Museum_Name','Rank','Rating','Tourist_Type','Tags','Address','Description','Longitude','Latitude','Photo_URL']]

museums
```

### Converting in lowercase

```python
museums['Tags'] = museums['Tags'].apply(lambda x:x.lower() )

museums.head()
```

## Applying vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=1000,stop_words='english')
```

```python
cv.fit_transform(museums['Tags']).toarray()
```

```python
cv.fit_transform(museums['Tags']).toarray().shape
```

```python
vectors = cv.fit_transform(museums['Tags']).toarray()

vectors
```

```python
vectors[0]
```

# Removing context which is useless and similar words - nltk

```python
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        ps.stem(i)
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)   
```

```python
museums['Tags'][1]
```

```python
stem('history museums, natural history museums, museumsimax theater, natural history, piggly wiggly store, miniature circus, traveling exhibits, tuesday afternoon, civil war, memphis history, yellow fever, free admission, interesting exhibits, gift shop, on tuesdays, spend an afternoon, all ages, few hours, whole family, on display, great place to visit, dinosaursthe pink palace museum is one of the largest facilities of its kind in the southeast. you can walk through a replica of the first self service grocery store in the country, clarence saunders piggly wiggly. explore the cultural and natural history of the mid south through exciting exhibits, dioramas, and audio visuals. trace memphis development from the time of spanish explorers through the civil war and the yellow fever epidemics. learn from the award winning medical exhibit how health care grew to be memphis largest industry.tennessee')
```

```python
museums['Tags']=museums['Tags'].apply(stem)
```

```python
vectors = cv.fit_transform(museums['Tags']).toarray()

vectors
```

### Here we can see the words are not repeated and it is also stem

Now we have to calculate the distance between the museums which are plots in 2d in vectors (744,1000)

We will find the angle between them without using distance formula(ucledean) bcz if the angle is small we can say that those 2 museum are close to each other finding distance in big dimentions is not a good way to do

## Using cosine similarity distance in sklearn

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vectors)
```

The similarity score must be 1 bcz its the first museums so you can see 1

```python
# For second museum
similarity[1]
```

# Recommender function for similar museum prediction

```python
def recommend(museum):
    museum_index= museums[museums['Museum_Name']==museum].index[0]
    distances = similarity[museum_index]
    museums_list = sorted(list(enumerate(distances)),reverse=True ,key=lambda x:x[1])[1:6]
    for i in museums_list:
        print(museums.iloc[i[0]].Museum_Name)
```

```python
recommend('Landis Valley Museum')
```
